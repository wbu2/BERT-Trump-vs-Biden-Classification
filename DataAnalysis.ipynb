```python
#Combine User's Tweets Into Single String
import pickle 

with open('joe-list.pkl', 'rb') as f:
    biden_tweets = pickle.load(f)

with open('trump-list.pkl', 'rb') as f:
    trump_tweets = pickle.load(f)

biden_combined = "".join(biden_tweets) 
trump_combined = "".join(trump_tweets)

#biden_combined
#trump_combined
```

```python
from wordcloud import WordCloud, STOPWORDS 
import matplotlib.pyplot as plt 


stopwords = set(STOPWORDS)

#remove 'amp' keyword that follows ampersands in text
stopwords.update(["amp"])

word_cloud = WordCloud(background_color ='white', stopwords = stopwords, max_words = 15)

tweets_list = [biden_combined, trump_combined]

for i in range(len(tweets_list)):
    word_cloud.generate(tweets_list[i])
    
    plt.figure(figsize = (8, 8), facecolor = None) 
    plt.imshow(word_cloud,interpolation="bilinear")
    plt.axis("off") 
    plt.tight_layout(pad = 0) 
    
    title = "Joe Biden" if i < 1 else "Donald Trump"
    plt.title(title, fontsize = 30)
    
plt.show()
```

Based on the most common words from Trump and Biden's timeline, it is pretty clear who each word cloud belongs to. Political affiliation isn't very obvious from looking at the word clouds themselves, however the word clouds show a clear distinction between Trump and Biden. Looking at Trump's word cloud, there are phrases he often uses such as "Fake News" or "great". This shows that this data is fit to be used for a binary classification problem between Trump and Biden.
